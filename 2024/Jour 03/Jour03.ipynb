Code 1*

```python
import re

with open('input.txt', 'r') as file:
    content = file.read()

pattern = r"mul\(\d+,\d+\)"
matches = re.findall(pattern, content)

print(matches)

somme = 0

def check(match) -> int:
    line = match.split('(')[1].split(')')[0].split(',')
    return line

def mult(line):
    return int(line[0])*int(line[1])

for match in matches:
    somme += mult(check(match))
print(somme)
```

Code 2*

```python
import re

somme = 0
etat = True
# Lire le contenu du fichier
with open('input.txt', 'r') as file:
    content = file.read()

# D√©finir l'expression r√©guli√®re pour trouver les occurrences de "mul(XXX,YYY)", "do()", et "don't()"
pattern = r'mul\(\d+,\d+\)|do\(\)|don\'t\(\)'

# Trouver toutes les correspondances
matches = re.findall(pattern, content)

def check(match) -> int:
    if(match[0] == 'm'):
        line = match.split('(')[1].split(')')[0].split(',')
        return line
    else:
        return [0, 0]

def mult(line):
    return int(line[0])*int(line[1])

# Afficher les correspondances trouv√©es
for match in matches:
    if etat == True:
        if match == "don't()":
            etat = False
        somme += mult(check(match))
    else:
        if match == "do()":
            etat = True

print (somme)
```
