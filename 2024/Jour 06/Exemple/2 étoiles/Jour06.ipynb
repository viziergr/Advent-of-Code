```python
with open('input.txt', 'r') as f:
    content = f.readlines()
    
def creationFile(f):
    """
    Pour chaque élément du fichier, on construit un tableau de la forme [[Id][0]]
    
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    content = []
    for ligne in f:
        line = []
        for char in ligne.strip():
            if(char == '.'):
                line.append('0')
            elif(char == '#'):
                line.append('1')
            elif(char == '>'):
                line.append('2')
            elif(char == 'v'):
                line.append('3')
            elif(char == '<'):
                line.append('4')
            elif(char == '^'):
                line.append('5')
        content.append(line)
    return content

content = creationFile(content)
print(content)
```

```python
def checkNext(tab):
    """
    Fonction qui regarde quel est le prochain élément en fonction de la direction de la garde
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] == '2':
                if tab[i][j+1] == '0':
                    return True
                else:
                    return False
            elif tab[i][j] == '4':
                if tab[i][j-1] == '0':
                    return True
                else:
                    return False
            elif tab[i][j] == '5':
                if tab[i-1][j] == '0':
                    return True
                else:
                    return False
            elif tab[i][j] == '3':
                if tab[i+1][j] == '0':
                    return True
                else:
                    return False
    return False
```

```python
def checkRotation(tab):
    """
    Fonction qui regarde si le garde doit tourner
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                garde_position = (i, j)
                direction = directions.index(tab[i][j])
                break
    
    # Regarder l'élément en face de la garde
    i, j = garde_position
    
    if direction == 0:  # droite
        if tab[i][j+1] == '1':
            return True
    elif direction == 1:  # bas
        if tab[i+1][j] == '1':
            return True
    elif direction == 2:  # gauche
        if tab[i][j-1] == '1':
            return True
    elif direction == 3:  # haut
        if tab[i-1][j] == '1':
            return True
```

```python
def rotation(tab):
    """
    Fonction qui regarde si le garde doit tourner
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                direction = directions.index(tab[i][j])
                
                if direction == 3:
                    tab[i][j] = directions[0]
                    
                else:
                    tab[i][j] = directions[direction+1]
                    
                break
    
    return tab
    
```

```python
def deplacement(tab):
    """
    On regarde le tableau donné en paramètre et on cherche où est situé la garde.
    En fonction de sa direction, on déplace le garde
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                garde_position = (i, j)
                direction = directions.index(tab[i][j])
                break

    # Déplacer la garde en fonction de la direction
    i, j = garde_position
    
    if direction == 0:  # droite
        tab[i][j], tab[i][j+1] = tab[i][j+1], tab[i][j]
        
    elif direction == 1:  # bas
        tab[i][j], tab[i+1][j] = tab[i+1][j], tab[i][j]
        
    elif direction == 2:  # gauche
        tab[i][j], tab[i][j-1] = tab[i][j-1], tab[i][j]
         
    elif direction == 3:  # haut
        tab[i][j], tab[i-1][j] = tab[i-1][j], tab[i][j]
        
    return tab
```
