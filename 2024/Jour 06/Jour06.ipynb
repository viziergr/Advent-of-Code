```python
with open('input.txt', 'r') as f:
    content = f.readlines()
    
    
def creationFile(f):
    """
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    content = []
    for ligne in f:
        line = []
        for char in ligne.strip():
            if(char == '.'):
                line.append('0')
            elif(char == '#'):
                line.append('1')
            elif(char == '>'):
                line.append('2')
            elif(char == 'v'):
                line.append('3')
            elif(char == '<'):
                line.append('4')
            elif(char == '^'):
                line.append('5')
        content.append(line)
    return content

content = creationFile(content)
print(content)
```

```python
def afficheMap(tab):
    print('-----------')
    for line in tab:
        print(line)
        print('\n')
    print('-----------')
```

```python
def checkNext(tab):
    """
    Fonction qui regarde quel est le prochain √©l√©ment en fonction de la direction de la garde
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] == '2':
                if tab[i][j+1] == '0':
                    return True
                else:
                    return False
            elif tab[i][j] == '4':
                if tab[i][j-1] == '0':
                    return True
                else:
                    return False
            elif tab[i][j] == '5':
                if tab[i-1][j] == '0':
                    return True
                else:
                    return False
            elif tab[i][j] == '3':
                if tab[i+1][j] == '0':
                    return True
                else:
                    return False
    return False
```

```python
def checkRotation(tab):
    """
    Fonction qui regarde si le garde doit tourner
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                garde_position = (i, j)
                direction = directions.index(tab[i][j])
                break
    
    # Regarder l'√©l√©ment en face de la garde
    i, j = garde_position
    
    if direction == 0:  # droite
        if tab[i][j+1] == '1':
            return True
    elif direction == 1:  # bas
        if tab[i+1][j] == '1':
            return True
    elif direction == 2:  # gauche
        if tab[i][j-1] == '1':
            return True
    elif direction == 3:  # haut
        if tab[i-1][j] == '1':
            return True
```

```python
def checkBord(tab):
    """
    Fonction qui regarde si la garde est en bord de map
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                garde_position = (i, j)
                direction = directions.index(tab[i][j])
                break
    
    i, j = garde_position
    if direction == 0:
        print(i,j)
        # Si la garde est vers la droite
        if j == len(tab[i])-1:
            return True
        
    if direction == 1:
        print(i,j)
        # Si la garde est vers le bas
        if i == len(tab)-1:
            return True
        
    if direction == 2:
        print(i,j)
        # Si la garde est vers la gauche
        if j == 0:
            return True
        
    if direction == 3:
        print(i,j)
        # Si la garde est vers le haut
        if j == 0:
            return True
    
    return False
```

```python
def mapping(set, tab):
    """
    Fonction qui ajoute les coordonn√©es par lesquelles la garde est pass√©e
    au set
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                garde_position = (i, j)
                break
    
    set.add(garde_position)
    return set
```

```python
def rotation(tab):
    """
    Fonction qui regarde si le garde doit tourner
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                direction = directions.index(tab[i][j])
                if direction == 3:
                    tab[i][j] = directions[0]
                else:
                    tab[i][j] = directions[direction+1]
                break
    
    return tab
    
```

```python
def deplacement(tab):
    """
    On regarde le tableau donn√© en param√®tre et on cherche o√π est situ√© la garde.
    En fonction de sa direction, on d√©place le garde
    Positions : 
    2 = vers la droite
    3 = vers le bas
    4 = vers la gauche
    5 = vers le haut
    """
    directions = ['2', '3', '4', '5']
    
    # Trouver la position de la garde
    for i in range(len(tab)):
        for j in range(len(tab[i])):
            if tab[i][j] in directions:
                garde_position = (i, j)
                direction = directions.index(tab[i][j])
                break

    # D√©placer la garde en fonction de la direction
    i, j = garde_position
    
    if direction == 0:  # droite
        tab[i][j], tab[i][j+1] = tab[i][j+1], tab[i][j]
    elif direction == 1:  # bas
        tab[i][j], tab[i+1][j] = tab[i+1][j], tab[i][j]
    elif direction == 2:  # gauche
        tab[i][j], tab[i][j-1] = tab[i][j-1], tab[i][j]   
    elif direction == 3:  # haut
        tab[i][j], tab[i-1][j] = tab[i-1][j], tab[i][j]
    return tab
```

```python
def main(content):
    setMapping = set([])
    while(checkNext(content)):
        mapping(setMapping, content)
        content = deplacement(content)
        if (checkBord(content)):
            break
        if(checkRotation(content)):
            content = rotation(content)
    return setMapping

map = main(content)

print(len(map) +1)
```
